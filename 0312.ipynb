# user define fun banavano 200+ A+ 180 A 170 B 150 C else F 

```python
def Grade(Total):
    t=Total
    if t>200:
        return "A+"
    elif t>180:
        return "A"
    elif t>170:
        return "B"
    elif t>150:
        return "C"
    else:
        return "F"
```

```python
df3["Grade"]=df3["Total"].apply(Grade)
```
