# All join in dbms

```python
df1=pd.DataFrame({"sub":["FSD","python","DM"],"fac":["nas","vha","mss"],"id":[1,2,4]})
df2=pd.DataFrame({"id":[1,2,3],"mark":[9,10,8]})
```

```python
df1
```

```python
df2
```

# There are 2 syntax for merge

# To merge atleast one col should be same

```python
pd.merge(df1,df2)
```

# by default id

```python
pd.merge(df1,df2,on="id")
```

# by default inner

```python
pd.merge(df1,df2,on="id",how="inner")
```

# outer

```python
pd.merge(df1,df2,on="id",how="outer")
```

# 2nd method

```python
df1.merge(df2)
```

```python
df1.merge(df2,on="id")
```

# df1 left and df2 right

# inner outer left and right join

# left

```python
pd.merge(df1,df2,on="id",how="left")
```

# right

```python
pd.merge(df1,df2,on="id",how="right")
```

# how="cross" give all possible combination that can be 

# version updated nathi etale error ave che

```python
pd.merge(df1,df2,how="cross")
```

# It works only for merging 2 tatble were there isn't coomon col

```python
df1=pd.DataFrame({"sub":["FSD","python","DM"],"fac":["nas","vha","mss"],"id_x":[1,2,4]})
df2=pd.DataFrame({"id_v":[1,2,3],"mark":[9,10,8]})
```

# Merge using different col name as 1

```python
df1.merge(df2,left_on="id_x",right_on="id_v")
```

# Concate using merge 

```python
df1.merge(df2,left_index=True,right_index=True)
```
