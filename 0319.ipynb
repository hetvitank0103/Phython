# With data text

```python
fig=px.bar(data_frame=df,x='post_type',y='likes',text_auto=True)
fig.show()
```

# With different colors for different group

```python
fig=px.bar(data_frame=df,x='post_type',y='likes',text_auto=True,color="platform")
fig.show()
```

# Change color colom catagery wise

```python
fig=px.bar(data_frame=df,x='post_type',y='likes',text_auto=True,color="platform",color_discrete_sequence=["pink","violet","cyan"])
fig.show()
```

```python
fig=px.bar(data_frame=df,x='post_type',y='likes',text_auto=True,color="platform",color_discrete_map={"Facebook":"pink","Instagram":"violet","Twitter":"cyan"})
fig.show()
```

# change height width

```python
fig=px.bar(data_frame=df,x='post_type',y='likes',text_auto=True,color="platform",height=700,width=2000,color_discrete_map={"Facebook":"pink","Instagram":"violet","Twitter":"cyan"})
fig.show()
```

# title

```python
fig=px.bar(data_frame=df,x='post_type',y='likes',text_auto=True,title="Platform wise post",color="platform",color_discrete_map={"Facebook":"pink","Instagram":"violet","Twitter":"cyan"})
fig.show()
```

# color ma numerical col

```python
fig=px.bar(data_frame=df,x='post_type',y='likes',text_auto=True,color="comments",color_discrete_sequence=["pink","violet","cyan"])
fig.show()
```

```python
fig=px.bar(data_frame=df,x='post_type',y='likes',text_auto=True,color="comments",hover_data=["post_day","sentiment_score"],color_discrete_sequence=["pink","violet","cyan"])
fig.show()
```

# list in y

```python
fig=px.bar(data_frame=df,x='post_type',y=['likes','comments','shares'],text_auto=True,color="platform",text="comments",color_discrete_sequence=["pink","violet","cyan"])
fig.show()
```

# by default barmode="stcak"

```python
fig=px.bar(data_frame=df,x='post_type',y=['likes','comments','shares'],text_auto=True,color="platform",text="comments",color_discrete_sequence=["pink","violet","cyan"],barmode="stack")
fig.show()
```

# by group

```python
fig=px.bar(data_frame=df,x='post_type',y=['likes','comments','shares'],text_auto=True,color="platform",color_discrete_sequence=["pink","violet","cyan"],barmode="group")
fig.show()
```

# display patern type with categorial col

```python
fig=px.bar(data_frame=df,x='post_type',y=['likes','comments','shares'],text_auto=True,color="platform",color_discrete_sequence=["pink","violet","cyan"],barmode="group",pattern_shape='sentiment_score')
fig.show()
```
